```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import undetected_chromedriver as uc
import time
```

```python
# 크롬 드라이버 열기
driver = uc.Chrome()
driver.get('https://music.youtube.com/')
```

```python
# xpath
xpath0_head="/html/body/ytmusic-app/ytmusic-app-layout/div[3]/ytmusic-browse-response/div[2]/ytmusic-section-list-renderer/div[2]/ytmusic-playlist-shelf-renderer/div[1]/ytmusic-responsive-list-item-renderer["
xpath0_tail="]/div[2]/div[1]/yt-formatted-string/a"
xpath1="/html/body/ytmusic-app/ytmusic-popup-container/tp-yt-iron-dropdown/div/ytmusic-menu-popup-renderer/tp-yt-paper-listbox/ytmusic-menu-navigation-item-renderer[2]/a/yt-formatted-string"
xpath2="/html/body/ytmusic-app/ytmusic-popup-container/tp-yt-iron-dropdown[2]/div/ytmusic-add-to-playlist-renderer/div[2]/div[2]/ytmusic-playlist-add-to-option-renderer[4]/button/div/yt-formatted-string[1]"
```

```python
# Element 접근하는 세 메소드
def get_element(xpath):
    try:
        time.sleep(0.2)
        return WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, xpath)))
    except StaleElementReferenceException:
        time.sleep(0.2)
        return get_element(xpath)

def get_element_text(xpath):
    try:
        time.sleep(0.2)
        txt=driver.find_element(By.XPATH,xpath).text
        return txt
    except:
        txt=""
        return txt

def get_element_click(xpath):
    try:
        time.sleep(0.2)
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, xpath))).click()
    except StaleElementReferenceException:
        time.sleep(0.2)
        get_element_click(xpath)
```

```python
errors=[]
scroll_idx=90

# 이전 플레이리스트는 총 431곡
for i in range(1,432):
    # 곡 xpath
    xpath0=xpath0_head+str(i)+xpath0_tail
    title=get_element_text(xpath0)
    
    try:    
        print(f"({i}/432) 진행중 : {title}")
        
        # 한 곡 픽셀만큼 스크롤 내리기
        driver.execute_script(f"window.scrollTo(0, {scroll_idx});")
        scroll_idx+=56
        
        # xpath0 우클릭, 곡정보 보기
        ActionChains(driver).context_click(get_element(xpath0)).perform()

        # xpath1 클릭, 재생 목록 추가 버튼
        get_element_click(xpath1)
    
        # xpath2 클릭, 해당 재생 목록 버튼
        get_element_click(xpath2)
    
    
    except Exception as e:
        errors.append(title)
        print("{} : {}".format(traceback.format_exc(), title))
```
